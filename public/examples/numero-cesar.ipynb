# El cifrado del César

Tu tarea es encriptar un mensaje utilizando el método de "la cifra del César". Este método consiste en desplazar cada letra del mensaje una cantidad fija de posiciones en el alfabeto.

### Reglas:
- Cada letra se desplaza hacia adelante en el alfabeto según el número de posiciones especificado.
- Si el desplazamiento supera la letra "z", el conteo continúa desde la letra "a".
- El método es insensible a mayúsculas y minúsculas: "A" se considera equivalente a "a", "B" a "b", y así sucesivamente. La salida siempre será en minúsculas.
- Los espacios y otros caracteres no alfabéticos permanecen inalterados en la encriptación.

### Ejemplos:

Desplazamiento de 2 posiciones:
Mensaje original: `HOLA`
Mensaje encriptado: `jqnc`

Desplazamiento de 3 posiciones:
Mensaje original: `hola mundo`
Mensaje encriptado: `krod pxqgr`

Desplazamiento de 5 posiciones:   
Mensaje original: `abc xyz`
Mensaje encriptado: `fgh cde`

Desplazamiento de 1 posición:
Mensaje original: `zapato`
Mensaje encriptado: `abqbpv`

### Notas adicionales:
- Si el alfabeto se termina antes de completar el desplazamiento necesario, se continúa desde el principio. Por ejemplo, con un desplazamiento de 2 posiciones, la letra "z" se convierte en "b".
- Asegúrate de que cualquier espacio o puntuación en el mensaje original permanezca inalterado en el mensaje encriptado.

Necesitamos que 97 <= letra + desplazamiento <= 122  
Para eso vamos a usar el operador módulo % 26 (las letras del alfabeto) para "dar la vuelta".  
Comenzar desplazando un solo carácter para verificar esto.   
Completar el código para que cualquier letra minúscula se desplace correctamente.

```java
int desplazamiento = 2;
char original = 'z';

// Lógica para que 'z' + 2 sea 'b', esto lo harían ellos. 
int base = (int) 'a';
char cifrado = (char) ( ( (original - base) + desplazamiento) % 26 + base);

cifrado
```
