# Manipulación de Strings en Java
## El Espejo (Invertir una Cadena)
En Java, las String son inmutables (no se pueden cambiar una vez creadas). 
### El Reto: Implementar invertir(String cadena).
Pista:  ¿podrías hacerlo manualmente con un bucle for invertido para entender la lógica?

```java
public String invertir(String cadena) {
    String resultado = "";
    for (int i = cadena.length() - 1; i >= 0; i--) {
        resultado += cadena.charAt(i);
    }
    return resultado;
}
```

```java
String cadena = "Hola Mundo";
System.out.println(invertir(cadena));
```

### El Detective de Palíndromos
Compararemos la palabra original con su versión invertida para saber si son palíndromos.

Cuidado: En Java, para comparar el contenido de dos String no usamos == (eso compara si son el mismo objeto en memoria), usamos **.equals()**.

```java
public boolean esPalindromo(String palabra) {
    String invertida = invertir(palabra);
    return palabra.equals(invertida);
}

// Prueba:
System.out.println(esPalindromo("radar")); // true
```

### El Desafío de la Limpieza (Normalización)
Para procesar frases como "La ruta nos aportó otro paso natural", necesitamos que Java ignore:

1. Mayúsculas: .toLowerCase()

1. Espacios y Signos.

1. Acentos.

Vamos a crear un método que "limpie" el texto antes de verificar si es palíndromo.

#### Veamos primero como dejar toda la frase en minúsculas (o en mayúsculas, si así lo prefieren)

```java
String conMayusculas = "La ruta nos aportó otro paso natural";
String todoEnMinusculas = conMayusculas.toLowerCase();
String todoEnMayusculas = conMayusculas.toUpperCase();
System.out.println(todoEnMinusculas);
System.out.println(todoEnMayusculas);
System.out.println(conMayusculas);
```

🤔 ¿Porqué al imprimir nuevamente la frase en la línea 6 la seguimos viendo en su forma original, si le hemos hecho dos transformaciones?

#### Ahora que tenemos toda la frase en minúsculas, quitemos los espacios y signos.

Investigamos el uso del método replace() de la class String

```java
System.out.println("Algo, otra cosa".replace(",",""));
//Hemos quitado la , (coma) teemplazandola por un string vacío. 
```

```java
String cadena = "¡Hola, mundo! @Java #Código.";
String aQuitar = " ,./?~!¡@#$%^&*()_+[]{}|;:";
String cadenaNormalizada = cadena;
for(int i=0; i<aQuitar.toCharArray().length; i++){
    cadenaNormalizada = cadenaNormalizada.replace(String.valueOf(aQuitar.charAt(i)),"");
}
System.out.println(cadenaNormalizada)
```

#### Ahora nos toca quitar los acentos

Vamos a usar la misma estrategia que para quitar los acentos. 

```java
String fraseConAcentos = "El pingüino Ramón comió ñame en el camión de Asunción.";
String fraseSinAcentos = fraseConAcentos;
String conAcentos = "áéíóúüñÁÉÍÓÚÜÑ";
String sinAcentos = "aeiouunAEIOUUN";

for(int i=0; i<conAcentos.length(); i++){
    fraseSinAcentos = fraseSinAcentos.replace(String.valueOf(conAcentos.charAt(i)),String.valueOf(sinAcentos.charAt(i)));
}

System.out.println(fraseSinAcentos);
```

#### ¡Estamos muy cerca de la solución!

Para poder descubrir frases palindrómicas, habiendo descubierto las estrategias para resolver cada parte, vamos a modularizar en funciones.

```java
String aMinusculas(String frase){
    return frase.toLowerCase();    
}

String sinEspaciosNiSignos(String frase){
    String aQuitar = " ,./?~!¡@#$%^&*()_+[]{}|;:'";
    String cadenaNormalizada = frase;
    for(int i=0; i<aQuitar.toCharArray().length; i++){
        cadenaNormalizada = cadenaNormalizada.replace(String.valueOf(aQuitar.charAt(i)),"");
    }
    return cadenaNormalizada;
}

String sinAcentos(String frase){
    String fraseSinAcentos = frase;
    // No necesitamos las mayúsculas
    String conAcentos = "áéíóúüñ";
    String sinAcentos = "aeiouun";

    for(int i=0; i<conAcentos.length(); i++){
        fraseSinAcentos = fraseSinAcentos.replace(conAcentos.charAt(i),sinAcentos.charAt(i));
    }

    return fraseSinAcentos;
}

String normalizar(String frase){
    frase = aMinusculas(frase);
    frase = sinEspaciosNiSignos(frase);
    frase = sinAcentos(frase);
    return frase;
}

boolean esPalindromo(String frase){
    frase = normalizar(frase);
    System.out.println(frase);
    return frase.equals(invertir(frase));
}

String poema = "'Amor azul'Ramera, de todo te di.Mariposa colosal, sí,yo de todo te di.Poda la rosa, Venus.El átomo como tal es un evasor alado.Pide, todo te doy: isla,sol, ocaso, pirámide.Todo te daré: mar, luz, aroma.";

esPalindromo(poema);
```
