## String.format y printf
Formato de texto con placeholders — requiere la implementacion completa de `java.util.Formatter`.

```java
String nombre = "Ada";
int edad = 36;
double pi = Math.PI;

System.out.println(String.format("Hola %s, tenes %d anios.", nombre, edad));
System.out.printf("Pi con 4 decimales: %.4f%n", pi);
System.out.printf("Numero grande: %,d%n", 1_000_000);
System.out.printf("%-20s | %5d%n", "Ada Lovelace", 1843);
System.out.printf("%-20s | %5d%n", "Alan Turing", 1912);
```

## Reflection
Inspeccionar clases en tiempo de ejecucion — listar metodos, campos y constructores.

```java
import java.lang.reflect.*;

class Vehiculo {
    private String marca;
    private int anio;
    
    Vehiculo(String marca, int anio) {
        this.marca = marca;
        this.anio = anio;
    }
    
    public String getMarca() { return marca; }
    public int getAnio() { return anio; }
    public String toString() { return marca + " (" + anio + ")"; }
}

var auto = new Vehiculo("Toyota", 2024);
System.out.println("Clase: " + auto.getClass().getSimpleName());
System.out.println("\nMetodos publicos:");
for (Method m : auto.getClass().getDeclaredMethods()) {
    System.out.printf("  %s %s()%n", m.getReturnType().getSimpleName(), m.getName());
}
System.out.println("\nCampos:");
for (Field f : auto.getClass().getDeclaredFields()) {
    System.out.printf("  %s %s%n", f.getType().getSimpleName(), f.getName());
}
```

## Records (Java 16+)
Clases de datos inmutables en una sola linea. El compilador genera `equals()`, `hashCode()`, `toString()` y accessors automaticamente.

```java
record Punto(double x, double y) {
    double distanciaA(Punto otro) {
        return Math.sqrt(Math.pow(x - otro.x, 2) + Math.pow(y - otro.y, 2));
    }
}

var origen = new Punto(0, 0);
var destino = new Punto(3, 4);

System.out.println(origen);
System.out.println(destino);
System.out.printf("Distancia: %.2f%n", origen.distanciaA(destino));
System.out.println("Iguales? " + origen.equals(new Punto(0, 0)));
```

## java.time (Java 8+)
API moderna de fecha y hora — mucho mas robusta que `Date` y `Calendar`.

```java
import java.time.*;
import java.time.format.DateTimeFormatter;

var hoy = LocalDate.now();
var nacimiento = LocalDate.of(1995, 3, 15);
var edad = java.time.Period.between(nacimiento, hoy);

System.out.println("Hoy: " + hoy);
System.out.println("Nacimiento: " + nacimiento);
System.out.printf("Edad: %d anios, %d meses%n", edad.getYears(), edad.getMonths());

var formatter = DateTimeFormatter.ofPattern("dd/MM/yyyy");
System.out.println("Formato local: " + hoy.format(formatter));

var ahora = LocalTime.now();
System.out.printf("Hora actual: %02d:%02d%n", ahora.getHour(), ahora.getMinute());
```

## Streams y Collectors
Procesamiento funcional de colecciones con lambdas.

```java
import java.util.*;
import java.util.stream.*;

var palabras = List.of("Java", "Python", "JavaScript", "Go", "Rust", "C", "TypeScript");

// Filtrar, transformar, recoger
var largas = palabras.stream().filter(p -> p.length() > 3).map(String::toUpperCase).sorted().collect(Collectors.joining(", "));
System.out.println("Lenguajes con mas de 3 letras: " + largas);

// Estadisticas
var stats = palabras.stream().mapToInt(String::length).summaryStatistics();
System.out.printf("Promedio de longitud: %.1f, Max: %d, Min: %d%n", stats.getAverage(), stats.getMax(), stats.getMin());

// Agrupar por longitud
var porLongitud = palabras.stream().collect(Collectors.groupingBy(String::length));
porLongitud.forEach((len, list) -> System.out.printf("  %d letras: %s%n", len, list));
```

## Pattern Matching con instanceof (Java 16+)
Elimina casts manuales — la variable se declara directamente en el `if`.

```java
import java.util.*;

interface Forma {
    double area();
}

record Circulo(double radio) implements Forma {
    public double area() { return Math.PI * radio * radio; }
}

record Rectangulo(double ancho, double alto) implements Forma {
    public double area() { return ancho * alto; }
}

String describir(Object obj) {
    if (obj instanceof Circulo c) {
        return String.format("Circulo de radio %.1f, area = %.2f", c.radio(), c.area());
    } else if (obj instanceof Rectangulo r) {
        return String.format("Rectangulo %sx%s, area = %.2f", r.ancho(), r.alto(), r.area());
    } else if (obj instanceof String s) {
        return "Texto: " + s;
    }
    return "Desconocido: " + obj;
}

var formas = List.of(new Circulo(5), new Rectangulo(3, 4), "no soy una forma", 42);
for (var f : formas) { System.out.println(describir(f)); }
```
