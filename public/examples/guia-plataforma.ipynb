## Tipos de celdas
Hay dos tipos de celdas:
- **Código**: ejecuta Java 17 completo (como la siguiente)
- **Markdown**: texto con formato (como esta que estás leyendo)

Usá los botones **+ Código** y **+ Markdown** entre las celdas para agregar nuevas.

```java
// Esta es una celda de código. Presioná el botón ▶ para ejecutarla.
System.out.println("¡Hola desde Java " + Runtime.version().feature() + "!");
```

## Ejecutar celdas
Hay varias formas de ejecutar código:

| Acción | Cómo |
|---|---|
| Ejecutar una celda | Botón **▶ Ejecutar** o **Shift+Enter** |
| Ejecutar y avanzar a la siguiente | **Ctrl+Enter** |
| Reiniciar y ejecutar todo | Menú del notebook (ícono **⋮**) > **Reiniciar y ejecutar todo** |
| Reiniciar sesión | Menú del notebook (ícono **⋮**) > **Reiniciar sesión** |

Ejecutá las siguientes dos celdas en orden para ver cómo las variables quedan en memoria y se pueden usar desde cualquier celda.

```java
// Las variables persisten entre celdas
String nombre = "Notebook";
int version = 1;
```

```java
// Esta celda usa las variables de la celda anterior
System.out.println(nombre + " v" + version);
```

## Editar celdas markdown
Hacé **doble clic** en cualquier celda de markdown para editarla. Al terminar:
- **Shift+Enter**: guarda y queda en la celda
- **Ctrl+Enter**: guarda y avanza a la siguiente
- **Escape**: sale sin avanzar

¡Probá haciendo doble clic en esta celda!

## Modo lectura
Desde **Configuración** (ícono de engranaje), cambiá el **Modo** a **Lectura**.

En este modo se ocultan todos los controles de edición y el notebook se ve limpio, ideal para presentar o compartir. También podés alternar con **Ctrl+E**.

## Acciones de celda
Cada celda de código tiene botones para **Subir**, **Bajar** y **Eliminar**.

También podés usar el menú del notebook (ícono **⋮** a la derecha) para cortar, copiar, pegar celdas, y deshacer eliminaciones.

## Pestañas
Podés trabajar con **múltiples notebooks a la vez** en pestañas separadas.

- Cada pestaña tiene su **propia sesión de Java**, las variables de una no afectan a otra
- Usá el botón **+** en la barra de pestañas para crear un notebook nuevo
- **Doble clic** en una pestaña para renombrarla
- Todo se **guarda automáticamente** en tu navegador

## Menú Archivo
Desde el menú **Archivo** podés:
- **Importar**: abrir un archivo .ipynb (también podés arrastrar y soltar)
- **Exportar**: descargar el notebook actual (o usar **Ctrl+S**)
- **Ejemplos**: abrir notebooks de ejemplo (como este)

## Atajos de teclado
Presioná **?** (fuera de una celda de código) para ver todos los atajos disponibles.

Los más útiles:

| Atajo | Acción |
|---|---|
| **Shift+Enter** | Ejecutar celda |
| **Ctrl+Enter** | Ejecutar y avanzar |
| **Ctrl+S** | Exportar notebook |
| **Ctrl+E** | Alternar modo lectura/edición |
| **Ctrl+Up/Down** | Navegar entre celdas |

Desde **Configuración** podés habilitar atajos adicionales para cortar, copiar, pegar y eliminar celdas.

## Configuración
El ícono de **engranaje** en la barra superior abre las opciones:
- **Tema**: Auto, Claro u Oscuro
- **Identación**: 2 o 4 espacios
- **Modo**: Edición o Lectura
- **Atajos generales**: On u Off (habilita cortar, copiar, pegar, eliminar y deshacer celdas)
