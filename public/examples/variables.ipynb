# Variables y Tipos
Tipos primitivos y operaciones basicas en Java.

```java
// Tipos primitivos
int edad = 25;
double pi = 3.14159;
boolean esJava = true;
char nota = 'A';

System.out.println("Edad: " + edad);
System.out.println("Pi: " + pi);
System.out.println("Es Java? " + esJava);
System.out.println("Nota: " + nota);
```

## Operaciones con Strings

```java
String hola = "Hola";
String mundo = "Mundo";
String combinado = hola + ", " + mundo + "!";

System.out.println(combinado);
System.out.println("Longitud: " + combinado.length());
System.out.println("Mayusculas: " + combinado.toUpperCase());
System.out.println("Caracter en 0: " + combinado.charAt(0));
```

## Inferencia de tipos con var (Java 10+)

```java
var mensaje = "Inferencia de tipos!";
var numero = 42;
var decimal = 3.14;

System.out.println(mensaje + " " + numero + " " + decimal);
```
