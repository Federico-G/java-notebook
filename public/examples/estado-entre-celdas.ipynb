# Estado entre Celdas
En este notebook, cada celda comparte la misma sesion de JShell. Las variables, clases y metodos que definas en una celda **estan disponibles en todas las siguientes**.

Para empezar limpio, usa **Reiniciar sesion** desde el menu.

## 1. Definir una clase

```java
class Cuenta {
    private String titular;
    private double saldo;
    
    Cuenta(String titular, double saldoInicial) {
        this.titular = titular;
        this.saldo = saldoInicial;
    }
    
    void depositar(double monto) {
        saldo += monto;
        System.out.printf("%s: +$%.2f (saldo: $%.2f)%n", titular, monto, saldo);
    }
    
    void retirar(double monto) {
        if (monto > saldo) {
            System.out.println("Fondos insuficientes!");
            return;
        }
        saldo -= monto;
        System.out.printf("%s: -$%.2f (saldo: $%.2f)%n", titular, monto, saldo);
    }
    
    double getSaldo() { return saldo; }
    String getTitular() { return titular; }
    
    public String toString() {
        return String.format("Cuenta[%s, $%.2f]", titular, saldo);
    }
}
```

## 2. Usar la clase definida arriba
La clase `Cuenta` fue definida en la celda anterior y sigue disponible.

```java
var alice = new Cuenta("Alice", 1000);
var bob = new Cuenta("Bob", 500);

alice.depositar(250);
bob.depositar(100);
alice.retirar(75);
```

## 3. El estado persiste
Las variables `alice` y `bob` siguen existiendo con sus saldos actualizados.

```java
System.out.println(alice);
System.out.println(bob);
System.out.printf("%nSaldo total: $%.2f%n", alice.getSaldo() + bob.getSaldo());
```

## 4. Agregar funcionalidad
Podemos definir nuevos metodos y clases que interactuen con lo que ya existe.

```java
void transferir(Cuenta origen, Cuenta destino, double monto) {
    if (origen.getSaldo() < monto) {
        System.out.println("Fondos insuficientes para transferir!");
        return;
    }
    origen.retirar(monto);
    destino.depositar(monto);
    System.out.println("Transferencia completada.");
}

transferir(alice, bob, 200);

System.out.println("\nEstado final:");
System.out.println(alice);
System.out.println(bob);
```

## 5. Colecciones que crecen
Podemos acumular datos entre celdas — ideal para construir paso a paso.

```java
import java.util.*;

var historial = new ArrayList<String>();
historial.add("Dia 1: Inicio del proyecto");
historial.add("Dia 2: Primer prototipo");
System.out.println("Historial tiene " + historial.size() + " entradas");
```

```java
// Seguimos agregando en la misma lista
historial.add("Dia 3: Testing");
historial.add("Dia 4: Deploy");

System.out.println("Historial completo:");
for (int i = 0; i < historial.size(); i++) {
    System.out.printf("  [%d] %s%n", i + 1, historial.get(i));
}
```

## 6. Redefinir clases y metodos
JShell permite **redefinir** clases y metodos sobre la marcha. Al redefinir una clase, las instancias existentes siguen usando la version anterior, pero las nuevas usan la version actualizada.

```java
record Saludo(String nombre) {
    String mensaje() { return "Hola, " + nombre + "!"; }
}

var s1 = new Saludo("Mundo");
System.out.println(s1.mensaje());
```

```java
// Redefinimos el record con un mensaje distinto
record Saludo(String nombre) {
    String mensaje() { return "Buenos dias, " + nombre + ". Bienvenido!"; }
}

var s2 = new Saludo("Java");
System.out.println(s2.mensaje());
```
