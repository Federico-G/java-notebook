# POO - Clases y Objetos
Clases, records e interfaces dentro del notebook.

Las clases definidas en celdas se convierten en **clases locales** dentro de `main()`.

Las celdas marcadas como **Global** comparten su codigo con las celdas siguientes.

```java
// Definir una clase simple
class Animal {
    String nombre;
    String sonido;

    Animal(String nombre, String sonido) {
        this.nombre = nombre;
        this.sonido = sonido;
    }

    void hablar() {
        System.out.println(nombre + " dice " + sonido + "!");
    }
}
```

```java
// Usar la clase de la celda anterior (que es Global)
var gato = new Animal("Gato", "Miau");
var perro = new Animal("Perro", "Guau");
gato.hablar();
perro.hablar();
```

## Records (Java 16+)
Clases de datos compactas.

```java
record Punto(double x, double y) {
    double distanciaA(Punto otro) {
        double dx = this.x - otro.x;
        double dy = this.y - otro.y;
        return Math.sqrt(dx * dx + dy * dy);
    }
}

var a = new Punto(0, 0);
var b = new Punto(3, 4);
System.out.println("Distancia: " + a.distanciaA(b));
```

## Interfaces y polimorfismo

```java
interface Figura {
    double area();
    String describir();
}

class Circulo implements Figura {
    double radio;
    Circulo(double r) { this.radio = r; }
    public double area() { return Math.PI * radio * radio; }
    public String describir() { return "Circulo(r=" + radio + ")"; }
}

class Rectangulo implements Figura {
    double ancho, alto;
    Rectangulo(double ancho, double alto) { this.ancho = ancho; this.alto = alto; }
    public double area() { return ancho * alto; }
    public String describir() { return "Rectangulo(" + ancho + "x" + alto + ")"; }
}

var figuras = List.<Figura>of(new Circulo(5), new Rectangulo(3, 4), new Circulo(2));
for (var f : figuras) {
    System.out.println(f.describir() + " -> area = " + Math.round(f.area() * 100) / 100.0);
}
```
